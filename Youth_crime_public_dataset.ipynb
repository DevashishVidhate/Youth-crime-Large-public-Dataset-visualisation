### Dropping the columns

#### Dropping columns street prefix ( null values = 2,92,463 which is approx around 3lakhs) and street suffix(null values = 3,00,662 which is approx around 3lakhss)-Because of the large sets of null values dropping the columns

#### Dropping column Zip Code, Agency, End Date/time, Sector, Beat, PRA, Address Number,Location from the dataset because they do not contribute to the analysis objectives.This will help streamline our data and focus on the relevant features.

```python
drop_Col = ['Zip Code','Agency','End_Date_Time','Street Prefix','Street Suffix','Sector','Beat','PRA','Address Number','Location']
df1 = df1.drop(columns = drop_Col)
```

```python
df1
```

### Replacing the Null Values

#### Replacing null values in columns - Crime Name1,Crime Name2, Crime Name3,Police District Name,Block Address,City,Street Name as Unknown

```python
null_Values_col = ['Crime Name1','Crime Name2','Crime Name3','Police District Name','Block Address','City','Street Name','Street Type']
```

```python
df1[null_Values_col]= df1[null_Values_col].fillna('Unknown')
```

#### Assuming that Start Date/time and Dispatch Date/time are same, Replace the null values as same as start Date/time.

```python
df1['Dispatch Date / Time'] = df1['Dispatch Date / Time'].fillna('df1[Start_Date_Time]')
```

##### Checking the null values

```python
df1.isnull().sum()
```

### Data Cleaning and Transformation

#### Separating the Date/time column

```python
df1['Dispatch Date'] = df1['Dispatch Date / Time'].str.split(' ').str[0]
df1['Dispatch Time'] = df1['Dispatch Date / Time'].str.split(' ').str[1]
df1['Start Date'] = df1['Start_Date_Time'].str.split(' ').str[0]
df1['Start Time'] = df1['Start_Date_Time'].str.split(' ').str[1]
```

```python
df1
```

#### Converting Start Date column to Datetime format to extract Day,Month,Year and Day of week

```python
df1['Start Date'] = pd.to_datetime(df1['Start Date'], errors='coerce')
```

```python
df1['day'] = df1['Start Date'].dt.day
df1['month'] = df1['Start Date'].dt.month
df1['year'] = df1['Start Date'].dt.year
```

```python
df1['Day_of_Week'] = df1['Start Date'].dt.day_name()
```

```python
df1
```

#### Extracting Seasons from month column

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
```

```python
df1['Season'] = df1['month'].apply(get_season)
```

```python
df1
```

#### Extracting hour from Start Time column

```python
df1['Start_Date_Time'] = pd.to_datetime(df1['Start_Date_Time'], format='%m/%d/%Y %I:%M:%S %p')
df1['hour'] = df1['Start_Date_Time'].dt.hour
```

```python
df1
```

### Filtered the data From year 2018-2022. This helps reduce clutter and ensures that our analysis is more efficient.Since, We need data from 2018-2022

```python
filtered_df = df1[(df1['Start Date'].dt.year >= 2018) & (df1['Start Date'].dt.year <= 2022)]
```

## Q1. Based on frequency, what are the top 10 crime categories in the dataset?

```python
#counting the number of crimes
crime_counts = df['Crime Name2'].value_counts().reset_index()
```

```python
crime_counts = crime_counts[~crime_counts['Crime Name2'].isin(['All Other Offenses'])]
```

```python
#Renaming the columns of crime_counts
crime_counts.columns = ['Crime Category' , 'Frequency']
```

```python
crime_counts
```

```python
#To identify the top 10 crime categories
top_10_crimes = crime_counts.head(10)
```

```python
top_10_crimes
```

```python
plt.figure(figsize=(10, 6))
bars = plt.barh(top_10_crimes['Crime Category'], top_10_crimes['Frequency'], color='lightcoral')
plt.xlabel('Frequency')
plt.ylabel('Crime Category')
plt.title('Top 10 Crime Categories', fontweight='bold')
plt.gca().invert_yaxis()  # To display the highest frequency on top
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             f'{bar.get_width()}',
             va='center', ha='left', fontsize=8 , fontweight='bold');
```

```python
top_10_crimes_percentages = (top_10_crimes['Frequency'] / top_10_crimes['Frequency'].sum()) * 100
```

```python
top_10_crimes_percentages
```

```python
#to label the values in donut chart
##labels = [f"{crime_name} ({count})" for crime_name, count in zip(top_10_Crimes.index, top_10_Crimes.values)]
labels = [f"{count/1000:.1f}K ({percent:.1f}%)" for count, percent in zip(top_10_crimes['Frequency'].values, top_10_crimes_percentages.values)]
```

```python
plt.figure(figsize = (12,8), dpi = 200)
plt.title("Top 10 Crimes", fontsize=16, fontweight='bold', color='darkblue', pad=20)
plt.pie(x = top_10_crimes['Frequency'].values, labels = labels , startangle = 140, wedgeprops = dict(width = 0.3))
plt.legend (top_10_crimes['Crime Category'].values, loc = 'center', frameon=False, title='Crime Categories' )
plt.axis('equal');
```

#### The graph provides a clear visual representation of the frequency of various crime categories, making it easy to compare them at a glance.

#### All Other Offenses stands out prominently, with a frequency of 54,154 incidents. This suggests that this category includes a diverse range of offenses, making it the most common type of crime in the dataset.

#### In contrast, categories like Theft From Motor Vehicle (18,877) and Simple Assault (15,798) are also significant but not as prevalent as All Other Offenses. These figures indicate that while these crimes are notable, they occur less frequently.

#### On the lower end of the spectrum, we see Identity Theft (7,887) and Theft from Building (7,862), which have the least occurrences in this dataset. This suggests that these types of crimes are less common compared to others.

#### Overall, this data can be invaluable for law enforcement and policymakers. By identifying which crime types are most prevalent, they can better allocate resources and develop effective prevention strategies.

## Q2.What trends and changes have occurred in crime rates from 2018 to 2022?

```python
# Group by 'Crime Name1' and 'Year' and count the occurrences
yearly_counts = df.groupby(['Crime Name1', 'year']).size().reset_index(name='Count')
```

```python
yearly_counts = yearly_counts[~yearly_counts['Crime Name1'].isin(['Other', 'Unknown'])]
```

```python
yearly_counts
```

```python
pivot_df= yearly_counts.pivot(index='year', columns='Crime Name1', values='Count').fillna(0)

# Plotting
pivot_df.plot(kind='bar', figsize=(12, 6))

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of Crimes/Offences Recorded')
plt.title('Total Recorded Incidents')
plt.title('Trends in Crime Rates (2018-2022)')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title='Category of Crime ')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_counts, x='year', y='Count', hue='Crime Name1', marker='o')

# Adding annotations for each point
for i in range(len(yearly_counts)):
    plt.annotate(yearly_counts['Count'].iloc[i],
                 (yearly_counts['year'].iloc[i], yearly_counts['Count'].iloc[i]),
                 textcoords="offset points",
                 xytext=(0,5),
                 ha='center')

# Adding labels and title
plt.xlabel('year',fontsize=14)
plt.ylabel('Number of Crimes/Offences Recorded', fontsize=14)
plt.title('Line Plot of Crimes Recorded by Year', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Category of Crime', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.grid()
plt.show()
```

#### The data indicates a downward trend in most crime categories over the years, particularly in Crime Against Society and Crime Against Person.
#### The spike in Crime Against Property in 2021 suggests a temporary increase, possibly due to specific events or changes in reporting practices.

```python
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x='City', y='Number of Crimes', data=top_cities_df, palette='viridis')

for index, row in top_cities_df.iterrows():
    barplot.text(row.name, row['Number of Crimes'], round(row['Number of Crimes'], 2), color='black', ha="center")

plt.xlabel('City')
plt.ylabel('Number of Crimes')
plt.title('Top 10 Cities with the highest Crime Rate')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

#### The data reveals significant variations in crime rates across different cities. Silver Spring has the highest number of reported crimes at 75,973, indicating a potential need for increased safety measures or community programs. Following are Gaithersburg and Rockville, with 32,289 and 30,788 crimes, respectively, suggesting they also face notable crime challenges. In contrast, cities like Montgomery Village (6,556) and Takoma Park (6,025) report much lower crime counts, which may reflect safer environments.
#### This dataset provides valuable insights into crime trends across different cities, highlighting areas that may require more attention and resources to enhance public safety.

## Q4.What differences exist in crime patterns between weekdays and weekends

```python
print(df.columns)
```

```python
crime_counts1 = df['Day_of_Week'].value_counts()
crime_counts1.columns = ['Day_of_Week' , 'Crime counts']
crime_counts1
```

```python
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
crime_counts1 = crime_counts1.reindex(ordered_days)
```

```python
crime_counts1 = crime_counts1.reset_index()
crime_counts1.columns = ['Day_of_Week', 'Counts']
```

```python
crime_counts1
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Counts', y='Day_of_Week', data=crime_counts1, palette='coolwarm')

for index, value in enumerate(crime_counts1['Counts']):
    plt.text(value, index, str(value))


plt.title('Crime Counts on Each Day of the Week')
plt.xlabel('Number of Crimes')
plt.ylabel('Day of the Week')
plt.tight_layout();
```

```python
plt.figure(figsize=(12, 6))

# Create a line plot
plt.plot(crime_counts1['Day_of_Week'], crime_counts1['Counts'], marker='o', linestyle='-', color='b')

# Set plot title and labels
plt.title('Crime Counts on Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')

# Show the plot
plt.tight_layout()
plt.show()
```

#### the visualizations effectively communicate the trend and distribution of crime counts each day of the week,particularly emphasizing the significance of Fridays.
####  The increase in crime on Fridays can be linked to heightened social activities, alcohol consumption, and late-night outings, which create opportunities for various offenses.
